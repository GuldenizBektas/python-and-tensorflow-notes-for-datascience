```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age = [10, 20, 30, 30, 40, 50, 60, 70, 75]
weight = [20, 60, 80, 85, 86, 87, 70, 90, 95]
```

```python
ages = np.array(age)
weights = np.array(weight)
```

```python
plt.plot(ages, weights, "r") # r, g, b, and y
plt.xlabel("Ages")
plt.ylabel("Weights")
plt.title("Weight-Age Relation")
```

```python
arr = np.linspace(0, 10, 20)
```

```python
arr
```

```python
arr2 = arr ** 3
```

```python
arr2
```

```python
plt.plot(arr, arr2, "g+-")
```

```python
plt.subplot(1, 2, 1)
plt.plot(arr, arr2, "r*-")
plt.subplot(1, 2, 2)
plt.plot(arr2, arr, "b--")
```

```python
fig = plt.figure()
figAxes = fig.add_axes([0.2, 0.2, 0.9, 0.9]) # left, right, width, height
figAxes.plot(arr, arr2, "g")
figAxes.set_xlabel("X")
figAxes.set_ylabel("Y")
figAxes.set_title("Grafik Title")
```

```python
figg = plt.figure()
axes1 = figg.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = figg.add_axes([0.2, 0.4, 0.3, 0.3])

axes1.plot(arr, arr2, "g")
axes1.set_xlabel("X ekseni")
axes1.set_ylabel("Y ekseni")
axes1.set_title("Ana Grafik Başlık")

axes2.plot(arr2, arr, "b")
axes2.set_xlabel("X ekseni")
axes2.set_ylabel("Y ekseni")
axes2.set_title("Küçük Grafik Başlık")
```

```python
(figg, axes) = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(arr, arr2, "g")
    ax.set_xlabel("X Ekseni")
    
plt.tight_layout()
```

```python
(figg, axes) = plt.subplots(2, 2)
axes[0][0].plot(arr, arr2, "b")
axes[0][1].plot(arr,arr2, "y")
axes[1][0].plot(arr, arr2, "b")
axes[1][1].plot(arr,arr2, "y")
```

```python
figg = plt.figure() #figsize, dpi

axes = figg.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(arr, arr ** 2, label="numpy array ** 2")
axes.plot(arr, arr ** 3, label = "numpy array ** 3")
axes.legend(loc=2)
```

```python
figg.savefig("figg.png", dpi=200)
```

```python
arr = np.linspace(0, 10, 20)
```

```python
arr
```

```python
arr2 = arr ** 2
```

```python
figg, axes = plt.subplots()
axes.plot(arr, arr2, color="#3A95A8", alpha=0.4)
axes.plot(arr2, arr, color="#C96F23")
```

```python
newfig, newax = plt.subplots()
newax.plot(arr, arr + 2, color="blue", linewidth=1.0)
newax.plot(arr, arr + 3, color="yellow", linewidth=1.)
newax.plot(arr, arr + 4, color="#C96F23", linestyle="-.") #:,--
newax.plot(arr, arr + 7, color="#000000", linestyle="-.", marker="o",
          markerfacecolor="red", markersize=8)
newax.plot(arr, arr + 8, color="#000000", linestyle="-.", marker="+",
          markerfacecolor="red", markersize=10)
```
